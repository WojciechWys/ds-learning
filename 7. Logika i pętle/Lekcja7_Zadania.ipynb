```python
# Zadanie 5

suma = 0
i = 1

while i <= 50:
    suma += i
    i += 1

# sprawdzenie wzorem
n = 50
wzor = n*(n+1) // 2

print("Suma z pętli:", suma)
print("Suma ze wzoru:", wzor)

if suma == wzor:
    print("Wynik poprawny")
else:
    print("Wynik niepoprawny")
```

```python
# Zadanie 9

wagi = [70, 85, 62, 95, 78]
wzrosty = [1.75, 1.80, 1.65, 1.78, 1.72]
for i in range(len(wagi)):
    bmi = wagi[i] / (wzrosty[i] ** 2)
    bmi = round(bmi, 2)
    if bmi < 18.5: kategoria = "nieodwaga"
    elif bmi < 25: kategoria = "norma"
    else: kategoria = "nadwaga"

    print(f"Pacjent {i+1}: BMI = {bmi}, Kategoria: {kategoria}")
```
