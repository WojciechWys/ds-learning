```python
blood_pressure = [120,125,118,190,122]
avg_blood_pressure = sum(blood_pressure) / len(blood_pressure)
print("Średnie ciśnienie wynosi:",avg_blood_pressure)

difference_squared = []

for x in blood_pressure:
    x = (x - avg_blood_pressure)**2
    difference_squared.append(x)
print(difference_squared)

sum_of_squares = sum(difference_squared)
print(sum_of_squares)

variance = sum_of_squares / (len(blood_pressure) - 1)
print(variance)

standard_deviation = variance ** 0.5
print(standard_deviation)
```
