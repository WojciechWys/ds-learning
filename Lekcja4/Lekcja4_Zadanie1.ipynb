```python
weight = 50
height = 1.75
BMI = weight / height**2
print(f"BMI: {BMI:.2f}")
if (BMI < 18.5):
    print("Niedowaga")
elif (BMI >= 18.5 and BMI < 24.9):
    print("Norma")
else:
    print("Nadwaga")
```

```python
#alternatywnie z inputem
weight = int(input("Podaj swoją wagę: "))
height = float(input("Podaj swój wzrost: "))
BMI = weight / height**2
print(f"BMI: {BMI:.2f}")
if (BMI < 18.5):
    print("Niedowaga")
elif (BMI >= 18.5 and BMI < 24.9):
    print("Norma")
else:
    print("Nadwaga")
```
