```python
# Zadanie 2

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mean = 100
std = 15

norm = stats.norm(loc=mean, scale=std)

# P(X < 85)
p_85 = norm.cdf(85)
# P(X > 120)
p_120 = 1 - norm.cdf(120)
# P(90 < X < 110)
p_110 = norm.cdf(110) - norm.cdf(90)

# Zakres osi X
x = np.linspace(mean - 4*std, mean + 4*std, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))

# Krzywa gęstości
plt.plot(x, y, linewidth=2)
plt.title("Normal distribution with highlighted probabilities")

# --- P(X < 85) ---
x_85 = np.linspace(x.min(), 85, 500)
plt.fill_between(x_85, norm.pdf(x_85), alpha=0.3)
plt.text(70, 0.002, f"P(X < 85) = {norm.cdf(85):.3f}")

# --- P(X > 120) ---
x_120 = np.linspace(120, x.max(), 500)
plt.fill_between(x_120, norm.pdf(x_120), alpha=0.3)
plt.text(122, 0.002, f"P(X > 120) = {1 - norm.cdf(120):.3f}")

# --- P(90 < X < 110) ---
x_mid = np.linspace(90, 110, 500)
plt.fill_between(x_mid, norm.pdf(x_mid), alpha=0.3)
plt.text(92, 0.015, f"P(90 < X < 110) = {norm.cdf(110) - norm.cdf(90):.3f}")

# Linie pomocnicze
plt.axvline(85, linestyle='--')
plt.axvline(120, linestyle='--')
plt.axvline(90, linestyle=':')
plt.axvline(110, linestyle=':')

plt.xlabel("X")
plt.ylabel("Probability Density")
plt.show()
```

```python
# Zadanie 3
import pandas as pd
from scipy.stats import shapiro

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
iris = pd.read_csv(url)

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for feature in features:
    data = iris[features]
    statistic, p_value = shapiro(data)
    print(f"Shapiro-Wilk test: statistic={statistic:.3f}, p_value={p_value:.3f}")
```
