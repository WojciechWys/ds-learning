```
# Zadanie 5

import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)

nan_sum = titanic.isnull().sum()
print(nan_sum)
missing_percentage = nan_sum / len(titanic) * 100
print(missing_percentage)
missing_only = missing_percentage[(missing_percentage > 10)]
print(missing_only)
```
