```
# Zadanie 1

import numpy as np

#Z pythonowym range
table1 = np.array(range(1,101))

#czysty numpy, szybszy
table2 = np.arange(1, 101)
suma = np.sum(table2)
srednia = np.mean(table2)
mediana = np.median(table2)
maksimum = np.max(table2)
minimum = np.min(table2)

print(suma, srednia, mediana, maksimum, minimum)
```

```
# Zadanie 2

before = np.array([100, 250, 75, 320, 180, 450, 90])
after = np.array([80, 200, 60, 280, 150, 400, 75])

differences = before - after
print(differences)

discounts = np.round(((differences / before) * 100), 2)
print(discounts)

average_discount = np.round(np.mean(discounts), 2)
print(average_discount)
```

```
# Zadanie 3

temps = np.array([15, 18, 22, 25, 23, 20, 17, 14, 16, 19, 21, 24])

print(temps[::2])
print(temps[-5:])
print(np.flip(temps, axis=0))
print(temps[temps > 20])
```

```
# Zadanie 5

sales = np.random.randint(10, 100, size=(4, 7))
print(sales)

print(f"Średnie per produkt:", np.mean(sales, axis=1))
print(f"Średnie per dzień: ", np.mean(sales, axis=0))
print(f"Dzień z najwyższą sprzedażą: ", np.argmax(np.sum(sales, axis=0)) + 1)
print(f"Produkt z najniższą średnią sprzedażą: ", np.argmin(np.mean(sales, axis=1)) + 1)
```

```
# Zadanie 6

grades = np.array([78, 45, 92, 67, 34, 88, 56, 91, 23, 85, 73, 95])

# Znadź wszystkie oceny >= 80
print(grades[grades >= 80])
# Znajdź wszystkie oceny <50 (oceny niedostateczne)
print(grades[grades < 50])
# Ile jest ocen w przedziale [60, 80]
suma = len(grades[(grades > 50) & (grades <= 80)])
print(suma)
# Zmień wszystkie oceny < 50 na 50
grades_modified = grades.copy()
grades_modified[grades < 50] = 50
print(grades_modified)
```
