```python
import tabulate

columnMin = 50
columnMax = 100
value = [75, 50, 100, 62.5]
normalized_value = [(x - columnMin)/(columnMax - columnMin) for x in value]
value_in_range = [(x >= 0 and x <= 1 ) for x  in normalized_value]
table_data = list(zip(value, normalized_value, value_in_range))
col_names = ["Wartość oryginalna", "Wartość znormalizowana", "Czy mieści się w zakresie [0,1]"]

print(f"\n{'='*80}")
print(f"{'Normalizacja Min-Max':^80}")
print(f"{'='*80}")
print(tabulate.tabulate(table_data, headers=col_names, numalign="center", stralign="center"))
```
