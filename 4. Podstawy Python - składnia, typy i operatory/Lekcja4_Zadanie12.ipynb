```python
import math
import datetime

sample_size = 50000
batch_size = 128
epochs = 50
time_for_batch = 0.15 # w sekundach

full_batches = sample_size // batch_size
leftover = sample_size % batch_size

batches_per_epoch = full_batches + (1 if leftover > 0 else 0)
# batches_per_epoch = math.ceil(sample_size / batch_size)
print("Liczba batchy na epokę wynosi:", batches_per_epoch)

total_time_seconds = time_for_batch * batches_per_epoch * epochs
print("Całkowity czas treningu w sekundach wynosi:",total_time_seconds)

hours_display = total_time_seconds // 3600
seconds_display = total_time_seconds % 60
total_time_seconds = total_time_seconds % 3600
minutes_display = total_time_seconds // 60

print("Szacowany czas: %d h %d m %d s"%(hours_display,minutes_display,seconds_display))
#nie wyświetla brakującej 0,5s
```
