```python
# Zadanie 1

class Produkt:
    def __init__(self, nazwa, cena, ilosc_w_magazynie):
        self.nazwa = nazwa
        self.cena = cena
        self.ilosc_w_magazynie = ilosc_w_magazynie

    def pokaz_info(self):
        print(f"Nazwa produktu: {self.nazwa}")
        print(f"Cena produktu: {self.cena}")
        print(f"Ilość na magazynie: {self.ilosc_w_magazynie}")

    def sprzedaj(self, ilosc):
        self.ilosc_w_magazynie = self.ilosc_w_magazynie - ilosc

    def dostawa(self, ilosc):
        self.ilosc_w_magazynie = self.ilosc_w_magazynie + ilosc

produkt1 = Produkt("laptop", 2500, 50)
produkt1.pokaz_info()

produkt1.sprzedaj(10)
produkt1.pokaz_info()

produkt1.dostawa(20)
produkt1.pokaz_info()
```

```python
# Zadanie 2

class Kalkulator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def dodaj(self):
        print(self.a + self.b)

    def odejmij(self):
        print(self.a - self.b)

    def pomnoz(self):
        print(self.a * self.b)

    def podziel(self):
        print(self.a / self.b)

obliczenie1 = Kalkulator(1, 2)
obliczenie1.dodaj()
obliczenie1.odejmij()
obliczenie1.pomnoz()
obliczenie1.podziel()
```

```python
# Zadanie 4

class Ksiazka:
    def __init__(self, tytul, autor, rok_wydania, liczba_stron):
        self.tytul = tytul
        self.autor = autor
        self.rok_wydania = rok_wydania
        self.liczba_stron = liczba_stron

    def __str__(self):
        return f"'{self.tytul}' by {self.autor} ({self.rok_wydania}), {self.liczba_stron} pages"

    def __len__(self):
        return self.liczba_stron

moja_ksiazka = Ksiazka("Python Programming", "John Doe", 2020, 350)
print(str(moja_ksiazka))
print(f"{len(moja_ksiazka)} stron")
```

```python
# Zadanie 6

class Pojazd:
    def __init__(self, marka, model, rok):
        self.marka = marka
        self.model = model
        self.rok = rok

    def pokaz_info(self):
        print(f"Pojazd: {self.marka} {self.model}, Rocznik: {self.rok}")

class Samochod(Pojazd):
    def __init__(self, marka, model, rok, liczba_drzwi):
        super().__init__(marka, model, rok)
        self.liczba_drzwi = liczba_drzwi
    def pokaz_info(self):
        print(f"Samochód: {self.marka} {self.model}, Rocznik: {self.rok}, Liczba drzwi: {self.liczba_drzwi}")

pojazd1 = Pojazd("Skoda", "Fabia", 2137)
pojazd1.pokaz_info()

samochod = Samochod("Mercedes", "Vito", 2016, 5)
samochod.pokaz_info()
```

```python
# Zadanie 10 - DO NAPISANIA SAMEMU

class Produkt:
    def __init__(self, nazwa, cena):
        self.nazwa = nazwa
        self.cena = cena

class Zamowienie:
    def __init__(self):
        self.lista_produktow = []

    def dodaj_produkt(self, produkt):
        self.lista_produktow.append(produkt)

    def laczna_wartosc(self):
        suma = 0
        for produkt in self.lista_produktow:
            suma += produkt.cena
        return suma

    def __str__(self):
        produkt_str = []
        for produkt in self.lista_produktow:
            produkt_str.append(produkt.nazwa)
        return produkt_str, self.laczna_wartosc()

lista_produktow_dict = {'laptop': 3000, 'telefon': 1500, 'słuchawki': 500}

my_zamowienie = Zamowienie()

for nazwa, cena in lista_produktow_dict.items():
    produkt = Produkt(nazwa, cena)
    my_zamowienie.dodaj_produkt(produkt)

my_zamowienie.pokaz_zamowienie()

```
