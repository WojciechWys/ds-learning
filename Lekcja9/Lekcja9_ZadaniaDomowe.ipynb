```
# Zadanie 3

class BankAccount:
    def __init__(self, wlasciciel, saldo):
        self.wlasciciel = wlasciciel
        self.saldo = saldo
    def wplac(self, kwota):
        self.saldo += kwota
    def wyplac(self, kwota):
        self.saldo -= kwota
    def pokaz_saldo(self):
        print(f"Właściciel: {self.wlasciciel}, saldo konta: {self.saldo}")

konto1 = BankAccount("Adam Małysz", 102)
konto2 = BankAccount("Orzeł", 7)

konto1.wplac(50)
konto1.pokaz_saldo()
konto2.wyplac(2)
konto2.pokaz_saldo()
```

```
# Zadanie 5

class Student:
    def __init__(self, imie, numer_indeksu):
        self.imie = imie
        self.numer_indeksu = numer_indeksu
        self.oceny = []
    def dodaj_ocene(self, ocena):
        self.oceny.append(ocena)
    def srednia(self):
        return sum(self.oceny)/len(self.oceny)
    def __str__(self):
        return f"Student {self.imie} o numerze indeksu {self.numer_indeksu} ma średnią {self.srednia()}"

student1 = Student("Hyzio", 115225)
student2 = Student("Dyzio", 332552)
student3 = Student("Zyzio", 552772)

student1.dodaj_ocene(5)
student1.dodaj_ocene(2)
student1.dodaj_ocene(3)

student2.dodaj_ocene(4)

student3.dodaj_ocene(3)

student1.srednia()
print(str(student1))
print(str(student2))
print(str(student3))
```

```
class Uczen:
    def __init__(self, imie, oceny=None):
        self.imie = imie
        self.oceny = oceny if oceny is not None else []

    def dodaj_ocene(self, ocena):
        self.oceny.append(ocena)

    def srednia(self):
        if not self.oceny:
            return 0
        return sum(self.oceny) / len(self.oceny)


class Klasa:
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self.uczniowie = []

    def dodaj_ucznia(self, uczen):
        self.uczniowie.append(uczen)

    def srednia_klasy(self):
        if not self.uczniowie:
            return 0
        wszystkie_oceny = []
        for uczen in self.uczniowie:
            wszystkie_oceny.extend(uczen.oceny)
        if not wszystkie_oceny:
            return 0
        return sum(wszystkie_oceny) / len(wszystkie_oceny)

    def najlepszy_uczen(self):
        if not self.uczniowie:
            return None
        return max(self.uczniowie, key=lambda uczen: uczen.srednia())


# ===== TEST =====

klasa = Klasa("3A")

u1 = Uczen("Ania", [5, 4, 5])
u2 = Uczen("Bartek", [3, 4, 4])
u3 = Uczen("Celina", [5, 5, 6])
u4 = Uczen("Daniel", [2, 3, 3])
u5 = Uczen("Ela", [4, 4, 5])

for uczen in [u1, u2, u3, u4, u5]:
    klasa.dodaj_ucznia(uczen)

print("Średnia klasy:", klasa.srednia_klasy())

najlepszy = klasa.najlepszy_uczen()
print("Najlepszy uczeń:", najlepszy.imie, "średnia:", najlepszy.srednia())
```
