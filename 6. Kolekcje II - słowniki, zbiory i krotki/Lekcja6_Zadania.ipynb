```python
# Zadanie 3
words = ['python', 'java', 'python', 'javascript', 'python', 'java',
'c++', 'python']

frequency = {}
for jezyk in words:
    frequency[jezyk] = frequency.get(jezyk, 0) + 1

print(frequency)
```

```python
# Zadanie 4
A = {1,2,3,4,5}
B = {4,5,6,7,8}

both_sets = A | B
print(both_sets)

intersect_sets = A & B
print(intersect_sets)

diff_sets = A - B
print(diff_sets)

symm_diff_sets = A ^ B
print(symm_diff_sets)
```

```python
# Zadanie 5
duplicates = [1, 2, 3, 2, 4, 1, 5, 3, 6, 4, 7]
print(duplicates)
unique_set = set(duplicates)
print(unique_set)
print(f"Długość nieobrobionej listy wynosi:", len(duplicates))
print(f"Długość zbioru unikalnego wynosi:", len(unique_set))
```
