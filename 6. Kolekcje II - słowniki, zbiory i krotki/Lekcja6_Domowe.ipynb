```python
# Zadanie 14

studenci = [
    ('Anna', 'Kowalska', 5.0, 4.5, 5.0),
    ('Jan', 'Nowak', 3.0, 3.5, 3.0),
    ('Maria', 'Zielińska', 4.0, 4.5, 5.0),
    ('Piotr', 'Wiśniewski', 2.0, 3.0, 2.5),
    ('Ewa', 'Wójcik', 5.0, 5.0, 4.5),
    ('Tomasz', 'Kamiński', 4.0, 3.5, 4.0)
]

grupy = {
    'wyróżniający': [],
    'dobry': [],
    'średni': []
}

for imie, nazwisko, o1, o2, o3 in studenci:
    srednia = round((o1 + o2 + o3) / 3, 2)
    student = (imie, nazwisko, srednia)

    if srednia >= 4.5:
        grupy['wyróżniający'].append(student)
    elif srednia >= 3.5:
        grupy['dobry'].append(student)
    else:
        grupy['średni'].append(student)

for kategoria, lista in grupy.items():
    print(f"\n{kategoria.upper()}:")
    for s in lista:
        print(s)
```

```python
# Zadanie 17

filmy_gatunki = {
    'Matrix': ['Akcja', 'Sci-Fi'],
    'Shrek': ['Komedia', 'Familijny', 'Animacja'],
    'Inception': ['Akcja', 'Sci-Fi', 'Thriller'],
    'Toy Story': ['Animacja', 'Familijny'],
    'Die Hard': ['Akcja', 'Thriller'],
    'The Hangover': ['Komedia']
}

gatunki_filmy = {}

for film, gatunki in filmy_gatunki.items():
    for gatunek in gatunki:
        gatunki_filmy.setdefault(gatunek, set()).add(film)

for gatunek, filmy in gatunki_filmy.items():
    print(f"{gatunek}: {filmy}")
```
