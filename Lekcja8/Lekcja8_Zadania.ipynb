```python
# Zadanie 1

def statystyki(liczby):
    suma = 0
    minimum = liczby[0]
    maksimum = liczby[0]

    for x in liczby:
        suma += x
        if x < minimum:
            minimum = x
        if x > maksimum:
            maksimum = x

    srednia = round(suma / len(liczby), 2)

    return (suma, srednia, minimum, maksimum)

oceny = [3.5, 4.0, 5.0, 3.0, 4.5, 5.0, 4.0, 3.5]

wynik = statystyki(oceny)

print("Suma:", wynik[0])
print("Średnia:", wynik[1])
print("Minimum:", wynik[2])
print("Maksimum:", wynik[3])
```

```python
# Zadanie 2

import pandas as pd

def celsius_to_fahrenheit(c):
    return (c * 9 / 5) + 32
def fahrenheit_to_celsius(f):
    return (f - 32) * 5 / 9
def celsius_to_kelvin(c):
    return c + 273.15

temperatury_c = [0, 10, 20, 30, 37, 100]

temperatury_f = []
temperatury_k = []

for c in temperatury_c:
    temperatury_f.append(round(celsius_to_fahrenheit(c), 2))
    temperatury_k.append(round(celsius_to_kelvin(c), 2))

df = pd.DataFrame({
    "*C": temperatury_c,
    "*F": temperatury_f,
    "*K": temperatury_k
})

print(df)
```

```python
# Zadanie 3

# List comprehension z warunkiem
parzyste = [x for x in range(1, 101) if x % 2 == 0]
print(parzyste[:10])

```

```python
# Zadanie 4

# List comprehension: tylko liczby nieparzyste, podnoszone do kwadratu
kwadraty_nieparzyste = [x**2 for x in range(1, 21) if x % 2 != 0]
print(kwadraty_nieparzyste)

```

```python
# Zadanie 5

produkty = [
    ("laptop", 2999, 4.5),
    ("mysz", 89, 4.2),
    ("klawiatura", 249, 4.8),
    ("monitor", 899, 4.3)
]

# Sortowanie według ceny (drugiego elementu tuple)
produkty_posortowane = sorted(produkty, key=lambda x: x[1])

print(produkty_posortowane)
```

```python
# Zadanie 6

def powitanie(imie, jezyk="polski"):
    if jezyk == "polski":
        print(f"Cześć, {imie}!")
    elif jezyk == "angielski":
        print(f"Hello, {imie}!")
    elif jezyk == "niemiecki":
        print(f"Hallo, {imie}!")
    else:
        print(f"Nieznany język. Witaj, {imie}!")

# Wywołania funkcji
powitanie("Anna")                   # Domyślnie polski
powitanie("Anna", "angielski")     # Angielski
powitanie("Anna", "niemiecki")     # Niemiecki
```
