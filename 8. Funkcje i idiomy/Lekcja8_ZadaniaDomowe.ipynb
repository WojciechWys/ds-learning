```python
# Zadanie 18

dane = list(range(1, 101))
k = 5

def k_fold_split(dane, k):
    if len(dane)%k != 0:
        print("Liczba danych powinna być podzielna przez liczbę części (foldów)")

    else:
        rozmiar_folda = len(dane)//k
        return [
            dane[i * rozmiar_folda : (i + 1) * rozmiar_folda]
            for i in range(k)
        ]

crossvalidation = k_fold_split(dane, k)

if crossvalidation is not None:
    n = 1
    for fold in crossvalidation:
        print(f"Fold {n}: {fold}")
        n += 1
```
